```python
import pandas as pd # –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ç–∞–±–ª–∏—á–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
import numpy as np # –¥–ª—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π —Å –º–∞—Å—Å–∏–≤–∞–º–∏
import math # –¥–ª—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π

# –°–æ–∑–¥–∞–µ–º –∫–ª–∞—Å—Å –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö
class MedicalDataClassifier:
    # –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –∫–ª–∞—Å—Å–∞, –∫–æ—Ç–æ—Ä—ã–π –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –æ–±—ä–µ–∫—Ç–∞
    def __init__(self, random_seed=42):
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ random_seed –≤ –∞—Ç—Ä–∏–±—É—Ç–µ –æ–±—ä–µ–∫—Ç–∞
        self.random_seed = random_seed
        # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º seed –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª numpy –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
        np.random.seed(random_seed)
    
    # –°–∏–≥–º–æ–∏–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    def logistic_activation(self, x):
        """–°–∏–≥–º–æ–∏–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è"""
        # –ï—Å–ª–∏ x –æ—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∏–π, –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 0 —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ exp(-x)
        if x < -100:
            # –í–æ–∑–≤—Ä–∞—â–∞–µ–º 0.0
            return 0.0
        # –ï—Å–ª–∏ x –æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π, –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 1 —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è
        elif x > 100:
            # –í–æ–∑–≤—Ä–∞—â–∞–µ–º 1.0
            return 1.0
        # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å–∏–≥–º–æ–∏–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏: 1 / (1 + e^(-x))
        return 1.0 / (1.0 + math.exp(-x))

# –°–æ–∑–¥–∞–µ–º —ç–∫–∑–µ–º–ø–ª—è—Ä –∫–ª–∞—Å—Å–∞ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
classifier = MedicalDataClassifier()
print("‚úì –ö–ª–∞—Å—Å MedicalDataClassifier —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def prepare_dataset(self, file_path):
    """–ó–ê–î–ê–ù–ò–ï 1: –ü—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥ –¥–∞–Ω–Ω—ã—Ö"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 1: –ü–†–ï–ü–†–û–¶–ï–°–°–ò–ù–ì –î–ê–ù–ù–´–•")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –ß–∏—Ç–∞–µ–º CSV —Ñ–∞–π–ª —Å –º–µ–¥–∏—Ü–∏–Ω—Å–∫–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –≤ DataFrame
    medical_data = pd.read_csv(file_path)
    # –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –¥–∞–Ω–Ω—ã—Ö, —á—Ç–æ–±—ã –Ω–µ –∏–∑–º–µ–Ω—è—Ç—å –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω—ã–π DataFrame
    processed_data = medical_data.copy()
    
    # –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –±–∏–Ω–∞—Ä–Ω—ã—Ö (–¥–≤–æ–∏—á–Ω—ã—Ö) –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    binary_mappings = {
        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ 'Sex': 'M' -> 1, 'F' -> 0
        'Sex': {'M': 1, 'F': 0},
        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ 'ExerciseAngina': 'Y' -> 1, 'N' -> 0
        'ExerciseAngina': {'Y': 1, 'N': 0}
    }
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É –∏ –µ–≥–æ —Å–ª–æ–≤–∞—Ä—é –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
    for column, mapping in binary_mappings.items():
        # –ü—Ä–∏–º–µ–Ω—è–µ–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ —Å–æ–≥–ª–∞—Å–Ω–æ —Å–ª–æ–≤–∞—Ä—é
        processed_data[column] = processed_data[column].map(mapping)
    
    # –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ —á–∏—Å–ª–∞
    category_encodings = {
        # –ö–æ–¥–∏—Ä—É–µ–º —Ç–∏–ø—ã –±–æ–ª–µ–π –≤ –≥—Ä—É–¥–∏: ATA->0, NAP->1, ASY->2, TA->3
        'ChestPainType': {'ATA': 0, 'NAP': 1, 'ASY': 2, 'TA': 3},
        # –ö–æ–¥–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≠–ö–ì: Normal->0, ST->1, LVH->2
        'RestingECG': {'Normal': 0, 'ST': 1, 'LVH': 2},
        # –ö–æ–¥–∏—Ä—É–µ–º –Ω–∞–∫–ª–æ–Ω —Å–µ–≥–º–µ–Ω—Ç–∞ ST: Up->0, Flat->1, Down->2
        'ST_Slope': {'Up': 0, 'Flat': 1, 'Down': 2}
    }
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–º—É —Å—Ç–æ–ª–±—Ü—É –∏ –µ–≥–æ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—é
    for column, encoding in category_encodings.items():
        # –ü—Ä–∏–º–µ–Ω—è–µ–º –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ —Å–æ–≥–ª–∞—Å–Ω–æ —Å–ª–æ–≤–∞—Ä—é
        processed_data[column] = processed_data[column].map(encoding)
    
    # –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π –æ–±—Ä–∞–±–æ—Ç–∫–∏
    numerical_features = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —á–∏—Å–ª–æ–≤–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É
    for feature in numerical_features:
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (NaN)
        if processed_data[feature].isnull().any():
            # –í—ã—á–∏—Å–ª—è–µ–º –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞
            median_val = processed_data[feature].median()
            # –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–µ–¥–∏–∞–Ω–æ–π
            processed_data[feature] = processed_data[feature].fillna(median_val)
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —á–∏—Å–ª–æ–≤–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É –¥–ª—è –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
    for feature in numerical_features:
        # –í—ã—á–∏—Å–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞
        mean_val = processed_data[feature].mean()
        # –í—ã—á–∏—Å–ª—è–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞
        std_val = processed_data[feature].std()
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –Ω–µ —Ä–∞–≤–Ω–æ –Ω—É–ª—é (–∏–∑–±–µ–≥–∞–µ–º –¥–µ–ª–µ–Ω–∏—è –Ω–∞ 0)
        if std_val > 0:
            # –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫: –≤—ã—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏ –¥–µ–ª–∏–º –Ω–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
            processed_data[feature] = (processed_data[feature] - mean_val) / std_val
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –ü–†–ï–ü–†–û–¶–ï–°–°–ò–ù–ì–ê
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± —É—Å–ø–µ—à–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–µ
    print("‚úì –î–∞–Ω–Ω—ã–µ —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω—ã –∏ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–º–µ—Ä –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤)
    print(f"‚úì –†–∞–∑–º–µ—Ä dataset: {len(processed_data)} –∑–∞–ø–∏—Å–µ–π, {len(processed_data.columns)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏
    print(f"‚úì –ü—Ä–∏–∑–Ω–∞–∫–∏ –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏: {list(processed_data.columns)}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ä–∞–∑–¥–µ–ª–∞ —Å —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–Ω—ã—Ö
    print(f"‚úì –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:")
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —Å—Ç–æ–ª–±—Ü—É –≤ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    for col in processed_data.columns:
        # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ: –∏–º—è, —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        print(f"  - {col}: {processed_data[col].dtype}, —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {processed_data[col].nunique()}")
    # –ü–µ—á–∞—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤–æ –≤—Å–µ–º –¥–∞—Ç–∞—Å–µ—Ç–µ
    print(f"‚úì –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è: {processed_data.isnull().sum().sum()}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ä–∞–∑–¥–µ–ª–∞ —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
    print(f"‚úì –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π:")
    # –°—á–∏—Ç–∞–µ–º –∏ –ø–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–¥–æ—Ä–æ–≤—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ (–∫–ª–∞—Å—Å 0)
    print(f"  - –ó–¥–æ—Ä–æ–≤—ã–µ (0): {(processed_data['HeartDisease'] == 0).sum()} –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤")
    # –°—á–∏—Ç–∞–µ–º –∏ –ø–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–ª—å–Ω—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ (–∫–ª–∞—Å—Å 1)
    print(f"  - –ë–æ–ª—å–Ω—ã–µ (1): {(processed_data['HeartDisease'] == 1).sum()} –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥–∞
    print("‚úì –ü—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–∏–Ω–≥ –∑–∞–≤–µ—Ä—à–µ–Ω —É—Å–ø–µ—à–Ω–æ!\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏–∑ –º–µ—Ç–æ–¥–∞
    return processed_data

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥ –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.prepare_dataset = prepare_dataset

# –í–´–ü–û–õ–ù–Ø–ï–ú –ü–†–ï–ü–†–û–¶–ï–°–°–ò–ù–ì –î–ê–ù–ù–´–•
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 1: –ü–†–ï–ü–†–û–¶–ï–°–°–ò–ù–ì –î–ê–ù–ù–´–•")
processed_data = classifier.prepare_dataset('heart.csv')
print("‚úì –ó–ê–î–ê–ù–ò–ï 1 –í–´–ü–û–õ–ù–ï–ù–û!")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def create_stratified_split(self, data, test_ratio=0.3, target_column='HeartDisease'):
    """–ó–ê–î–ê–ù–ò–ï 2: –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–µ –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 2: –†–ê–ó–î–ï–õ–ï–ù–ò–ï –î–ê–ù–ù–´–•")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –í—ã–±–∏—Ä–∞–µ–º –∏–∑ –¥–∞–Ω–Ω—ã—Ö –≤—Å–µ—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –±–æ–ª–µ–∑–Ω–∏ (–∫–ª–∞—Å—Å 0)
    healthy_patients = data[data[target_column] == 0]
    # –í—ã–±–∏—Ä–∞–µ–º –∏–∑ –¥–∞–Ω–Ω—ã—Ö –≤—Å–µ—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ —Å –±–æ–ª–µ–∑–Ω—å—é (–∫–ª–∞—Å—Å 1)
    sick_patients = data[data[target_column] == 1]
    
    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–µ–º –∑–¥–æ—Ä–æ–≤—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ (frac=1 –æ–∑–Ω–∞—á–∞–µ—Ç 100% –¥–∞–Ω–Ω—ã—Ö)
    healthy_patients = healthy_patients.sample(frac=1, random_state=self.random_seed)
    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–µ–º –±–æ–ª—å–Ω—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ
    sick_patients = sick_patients.sample(frac=1, random_state=self.random_seed)
    
    # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–¥–æ—Ä–æ–≤—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ (30% –æ—Ç –≤—Å–µ—Ö –∑–¥–æ—Ä–æ–≤—ã—Ö)
    test_healthy_count = int(len(healthy_patients) * test_ratio)
    # –í—ãcomputing –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–ª—å–Ω—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ (30% –æ—Ç –≤—Å–µ—Ö –±–æ–ª—å–Ω—ã—Ö)
    test_sick_count = int(len(sick_patients) * test_ratio)
    
    # –°–æ–∑–¥–∞–µ–º —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É: –æ–±—ä–µ–¥–∏–Ω—è–µ–º –ø–µ—Ä–≤—ã—Ö test_healthy_count –∑–¥–æ—Ä–æ–≤—ã—Ö –∏ test_sick_count –±–æ–ª—å–Ω—ã—Ö
    test_set = pd.concat([
        # –ë–µ—Ä–µ–º –ø–µ—Ä–≤—ã—Ö test_healthy_count –∑–¥–æ—Ä–æ–≤—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ (—Å—Ä–µ–∑ –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ test_healthy_count)
        healthy_patients.iloc[:test_healthy_count],
        # –ë–µ—Ä–µ–º –ø–µ—Ä–≤—ã—Ö test_sick_count –±–æ–ª—å–Ω—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤
        sick_patients.iloc[:test_sick_count]
    ])
    
    # –°–æ–∑–¥–∞–µ–º –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É: –æ–±—ä–µ–¥–∏–Ω—è–µ–º –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤
    train_set = pd.concat([
        # –ë–µ—Ä–µ–º –≤—Å–µ—Ö –∑–¥–æ—Ä–æ–≤—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –Ω–∞—á–∏–Ω–∞—è —Å test_healthy_count (—Å—Ä–µ–∑ –æ—Ç test_healthy_count –¥–æ –∫–æ–Ω—Ü–∞)
        healthy_patients.iloc[test_healthy_count:],
        # –ë–µ—Ä–µ–º –≤—Å–µ—Ö –±–æ–ª—å–Ω—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤ –Ω–∞—á–∏–Ω–∞—è —Å test_sick_count
        sick_patients.iloc[test_sick_count:]
    ])
    
    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–µ–º –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ
    train_set = train_set.sample(frac=1, random_state=self.random_seed)
    # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–µ–º —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ
    test_set = test_set.sample(frac=1, random_state=self.random_seed)
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –†–ê–ó–î–ï–õ–ï–ù–ò–Ø
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± —É—Å–ø–µ—à–Ω–æ–º —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–∏
    print("‚úì –°—Ç—Ä–∞—Ç–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ —É—Å–ø–µ—à–Ω–æ")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ –∏ –µ–µ –¥–æ–ª—é –æ—Ç –æ–±—â–µ–≥–æ –æ–±—ä–µ–º–∞ –¥–∞–Ω–Ω—ã—Ö
    print(f"‚úì –û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞: {len(train_set)} –∑–∞–ø–∏—Å–µ–π ({len(train_set)/len(data):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –∏ –µ–µ –¥–æ–ª—é –æ—Ç –æ–±—â–µ–≥–æ –æ–±—ä–µ–º–∞ –¥–∞–Ω–Ω—ã—Ö
    print(f"‚úì –¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞: {len(test_set)} –∑–∞–ø–∏—Å–µ–π ({len(test_set)/len(data):.1%})")
    
    # –ê–Ω–∞–ª–∏–∑ –±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤
    # –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–¥–æ—Ä–æ–≤—ã—Ö –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    train_healthy = (train_set[target_column] == 0).sum()
    # –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–ª—å–Ω—ã—Ö –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    train_sick = (train_set[target_column] == 1).sum()
    # –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–¥–æ—Ä–æ–≤—ã—Ö –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    test_healthy = (test_set[target_column] == 0).sum()
    # –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–ª—å–Ω—ã—Ö –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    test_sick = (test_set[target_column] == 1).sum()
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –±–∞–ª–∞–Ω—Å–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"‚úì –ë–∞–ª–∞–Ω—Å –∫–ª–∞—Å—Å–æ–≤ –≤ –û–ë–£–ß–ê–Æ–©–ï–ô –≤—ã–±–æ—Ä–∫–µ:")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –∑–¥–æ—Ä–æ–≤—ã—Ö –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"  - –ó–¥–æ—Ä–æ–≤—ã–µ: {train_healthy} ({train_healthy/len(train_set):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"  - –ë–æ–ª—å–Ω—ã–µ: {train_sick} ({train_sick/len(train_set):.1%})")
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –±–∞–ª–∞–Ω—Å–µ –∫–ª–∞—Å—Å–æ–≤ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"‚úì –ë–∞–ª–∞–Ω—Å –∫–ª–∞—Å—Å–æ–≤ –≤ –¢–ï–°–¢–û–í–û–ô –≤—ã–±–æ—Ä–∫–µ:")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –∑–¥–æ—Ä–æ–≤—ã—Ö –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"  - –ó–¥–æ—Ä–æ–≤—ã–µ: {test_healthy} ({test_healthy/len(test_set):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"  - –ë–æ–ª—å–Ω—ã–µ: {test_sick} ({test_sick/len(test_set):.1%})")
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –ø—Ä–æ–ø–æ—Ä—Ü–∏–π –º–µ–∂–¥—É –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏ –≤—ã–±–æ—Ä–∫–∞–º–∏
    # –í—ã—á–∏—Å–ª—è–µ–º –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    original_ratio = (data[target_column] == 1).mean()
    # –í—ã—á–∏—Å–ª—è–µ–º –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    train_ratio = (train_set[target_column] == 1).mean()
    # –í—ã—á–∏—Å–ª—è–µ–º –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    test_ratio_val = (test_set[target_column] == 1).mean()
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–∏ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π
    print(f"‚úì –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π:")
    # –ü–µ—á–∞—Ç–∞–µ–º –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    print(f"  - –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: {original_ratio:.1%} –±–æ–ª—å–Ω—ã—Ö")
    # –ü–µ—á–∞—Ç–∞–µ–º –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"  - –û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞: {train_ratio:.1%} –±–æ–ª—å–Ω—ã—Ö")
    # –ü–µ—á–∞—Ç–∞–µ–º –¥–æ–ª—é –±–æ–ª—å–Ω—ã—Ö –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
    print(f"  - –¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞: {test_ratio_val:.1%} –±–æ–ª—å–Ω—ã—Ö")
    # –ü–µ—á–∞—Ç–∞–µ–º –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–æ–ª–∏ –±–æ–ª—å–Ω—ã—Ö –º–µ–∂–¥—É –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–æ–π
    print(f"  - –û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {abs(original_ratio - train_ratio):.3f}\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –º–µ—Ç–æ–¥–∞
    return train_set, test_set

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥ –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.create_stratified_split = create_stratified_split

# –í–´–ü–û–õ–ù–Ø–ï–ú –†–ê–ó–î–ï–õ–ï–ù–ò–ï –î–ê–ù–ù–´–•
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 2: –†–ê–ó–î–ï–õ–ï–ù–ò–ï –î–ê–ù–ù–´–•")
train_set, test_set = classifier.create_stratified_split(processed_data)
print("‚úì –ó–ê–î–ê–ù–ò–ï 2 –í–´–ü–û–õ–ù–ï–ù–û!")

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞–¥–∞–Ω–∏–π
X_train = train_set.drop('HeartDisease', axis=1).values
y_train = train_set['HeartDisease'].values
X_test = test_set.drop('HeartDisease', axis=1).values
y_test = test_set['HeartDisease'].values
print(f"‚úì –î–∞–Ω–Ω—ã–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω—ã: X_train {X_train.shape}, y_train {y_train.shape}, X_test {X_test.shape}, y_test {y_test.shape}")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def compute_gradient(self, X, y, coefficients):
    """–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞"""
    # –°–æ–∑–¥–∞–µ–º –º–∞—Å—Å–∏–≤ –Ω—É–ª–µ–π —Ç–æ–π –∂–µ –¥–ª–∏–Ω—ã —á—Ç–æ –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞
    gradient = np.zeros_like(coefficients)
    # –ü–æ–ª—É—á–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞–∑—Ü–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö
    sample_count = len(X)
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É –æ–±—Ä–∞–∑—Ü—É –¥–∞–Ω–Ω—ã—Ö
    for i in range(sample_count):
        # –í—ã—á–∏—Å–ª—è–µ–º –ª–∏–Ω–µ–π–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ (z = X * beta)
        linear_combination = np.dot(X[i], coefficients)
        # –ü—Ä–∏–º–µ–Ω—è–µ–º —Å–∏–≥–º–æ–∏–¥–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(y=1|x)
        prediction = self.logistic_activation(linear_combination)
        # –í—ã—á–∏—Å–ª—è–µ–º –æ—à–∏–±–∫—É: —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É —Ä–µ–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ–º
        error = y[i] - prediction
        
        # –û–±–Ω–æ–≤–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞
        for j in range(len(coefficients)):
            # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º j-–π —ç–ª–µ–º–µ–Ω—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ –Ω–∞ –æ—à–∏–±–∫—É * j-–π –ø—Ä–∏–∑–Ω–∞–∫ (–ø—Ä–∞–≤–∏–ª–æ —Ü–µ–ø–æ—á–∫–∏)
            gradient[j] += error * X[i][j]
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç (–¥–µ–ª–∏–º –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞–∑—Ü–æ–≤)
    return gradient / sample_count

def calculate_likelihood(self, X, y, coefficients):
    """–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è"""
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—É–º–º–∞—Ä–Ω–æ–µ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏–µ
    total_likelihood = 0.0
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –≤—Å–µ–º –æ–±—Ä–∞–∑—Ü–∞–º
    for i in range(len(X)):
        # –í—ã—á–∏—Å–ª—è–µ–º –ª–∏–Ω–µ–π–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –¥–ª—è i-–≥–æ –æ–±—Ä–∞–∑—Ü–∞
        z = np.dot(X[i], coefficients)
        # –í—ã—á–∏—Å–ª—è–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å –ø–æ–º–æ—â—å—é —Å–∏–≥–º–æ–∏–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
        prob = self.logistic_activation(z)
        # –ï—Å–ª–∏ —Ä–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å = 1
        if y[i] == 1:
            # –î–æ–±–∞–≤–ª—è–µ–º –ª–æ–≥–∞—Ä–∏—Ñ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ (–∑–∞—â–∏—Ç–∞ –æ—Ç log(0) —Å –ø–æ–º–æ—â—å—é max(prob, 1e-8))
            total_likelihood += math.log(max(prob, 1e-8))
        else:
            # –ï—Å–ª–∏ —Ä–µ–∞–ª—å–Ω—ã–π –∫–ª–∞—Å—Å = 0, –¥–æ–±–∞–≤–ª—è–µ–º –ª–æ–≥–∞—Ä–∏—Ñ–º (1 - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å)
            total_likelihood += math.log(max(1 - prob, 1e-8))
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—É–º–º–∞—Ä–Ω–æ–µ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏–µ
    return total_likelihood

def train_logistic_model(self, X_train, y_train, learning_rate=0.1, max_iterations=3000):
    """–ó–ê–î–ê–ù–ò–ï 3: –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 3: –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–ê–Ø –†–ï–ì–†–ï–°–°–ò–Ø - –û–ë–£–ß–ï–ù–ò–ï")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –æ–±—É—á–∞—é—â–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ numpy –º–∞—Å—Å–∏–≤ —Ç–∏–ø–∞ float –¥–ª—è —á–∏—Å–ª–µ–Ω–Ω–æ–π —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏
    X_array = np.array(X_train, dtype=float)
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ü–µ–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ numpy –º–∞—Å—Å–∏–≤ —Ç–∏–ø–∞ float
    y_array = np.array(y_train, dtype=float)
    # –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü –∏–∑ –µ–¥–∏–Ω–∏—Ü –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞ (intercept) - –∫–æ–Ω–∫–∞—Ç–µ–Ω–∏—Ä—É–µ–º –µ–¥–∏–Ω–∏—á–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –¥–∞–Ω–Ω—ã–º–∏
    X_with_bias = np.column_stack([np.ones(len(X_array)), X_array])
    
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Å–ª—É—á–∞–π–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –æ—Ç -0.1 –¥–æ 0.1
    model_coefficients = np.random.uniform(-0.1, 0.1, X_with_bias.shape[1])
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –Ω–∞—á–∞–ª–µ –æ–±—É—á–µ–Ω–∏—è
    print("–ù–∞—á–∞–ª–æ –æ–±—É—á–µ–Ω–∏—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏...")
    # –ü–µ—á–∞—Ç–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–±—É—á–µ–Ω–∏—è: —Å–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π
    print(f"–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–±—É—á–µ–Ω–∏—è: learning_rate={learning_rate}, max_iterations={max_iterations}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞–∑—Ü–æ–≤ –∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    print(f"–†–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö: {X_with_bias.shape[0]} samples, {X_with_bias.shape[1]} features")
    
    # –¶–∏–∫–ª –æ–±—É—á–µ–Ω–∏—è - –ø—Ä–æ—Ö–æ–¥–∏–º max_iterations —Ä–∞–∑ (–≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π –ø–æ–¥—ä–µ–º)
    for iteration in range(max_iterations):
        # –í—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç –Ω–∞ —Ç–µ–∫—É—â–µ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏
        gradient = self.compute_gradient(X_with_bias, y_array, model_coefficients)
        # –û–±–Ω–æ–≤–ª—è–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã: –Ω–æ–≤—ã–µ = —Å—Ç–∞—Ä—ã–µ + —Å–∫–æ—Ä–æ—Å—Ç—å_–æ–±—É—á–µ–Ω–∏—è * –≥—Ä–∞–¥–∏–µ–Ω—Ç
        model_coefficients += learning_rate * gradient
        
        # –ö–∞–∂–¥—ã–µ 600 –∏—Ç–µ—Ä–∞—Ü–∏–π –≤—ã–≤–æ–¥–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø—Ä–æ—Ü–µ—Å—Å–µ –æ–±—É—á–µ–Ω–∏—è
        if iteration % 600 == 0:
            # –í—ã—á–∏—Å–ª—è–µ–º —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è
            current_likelihood = self.calculate_likelihood(X_with_bias, y_array, model_coefficients)
            # –ü–µ—á–∞—Ç–∞–µ–º –Ω–æ–º–µ—Ä –∏—Ç–µ—Ä–∞—Ü–∏–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è
            print(f"–ò—Ç–µ—Ä–∞—Ü–∏—è {iteration}: –§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è = {current_likelihood:.4f}")
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –û–ë–£–ß–ï–ù–ò–Ø
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –æ–±—É—á–µ–Ω–∏—è
    print("\n‚úì –û–±—É—á–µ–Ω–∏–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ä–∞–∑–¥–µ–ª–∞ —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏
    print("‚úì –í—ã—á–∏—Å–ª–µ–Ω–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã:")
    # –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –∏–º–µ–Ω –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: Intercept (—Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω) –∏ Feature_1, Feature_2, ...
    feature_names = ['Intercept'] + [f'Feature_{i+1}' for i in range(X_train.shape[1])]
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—É –∏ –µ–≥–æ –∏–º–µ–Ω–∏
    for name, coef in zip(feature_names, model_coefficients):
        # –ü–µ—á–∞—Ç–∞–µ–º –∏–º—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞, –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—é (—É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç/—É–º–µ–Ω—å—à–∞–µ—Ç —Ä–∏—Å–∫)
        print(f"  {name:15}: {coef:10.4f} {'(—É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç —Ä–∏—Å–∫)' if coef > 0 else '(—É–º–µ–Ω—å—à–∞–µ—Ç —Ä–∏—Å–∫)'}")
    
    # –ü–µ—á–∞—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –æ–±—É—á–µ–Ω–∏—è
    print(f"‚úì –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π: {max_iterations}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è
    print(f"‚úì Final likelihood: {current_likelihood:.4f}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± —É—Å–ø–µ—à–Ω–æ–º –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
    print("‚úì –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —É—Å–ø–µ—à–Ω–æ –≤—ã—á–∏—Å–ª–µ–Ω—ã!\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ–±—É—á–µ–Ω–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏–∑ –º–µ—Ç–æ–¥–∞
    return model_coefficients

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥—ã –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.compute_gradient = compute_gradient
MedicalDataClassifier.calculate_likelihood = calculate_likelihood
MedicalDataClassifier.train_logistic_model = train_logistic_model

# –í–´–ü–û–õ–ù–Ø–ï–ú –û–ë–£–ß–ï–ù–ò–ï –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–û–ô –†–ï–ì–†–ï–°–°–ò–ò
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 3: –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–ê–Ø –†–ï–ì–†–ï–°–°–ò–Ø - –í–´–ß–ò–°–õ–ï–ù–ò–ï –ö–û–≠–§–§–ò–¶–ò–ï–ù–¢–û–í")
lr_coefficients = classifier.train_logistic_model(X_train, y_train)
print("‚úì –ó–ê–î–ê–ù–ò–ï 3 –í–´–ü–û–õ–ù–ï–ù–û!")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def predict_logistic(self, X_test, coefficients, decision_threshold=0.5):
    """–ó–ê–î–ê–ù–ò–ï 4: –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 4: –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–ê–Ø –†–ï–ì–†–ï–°–°–ò–Ø - –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ï")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –º–µ—Ç–æ–∫
    predictions = []
    # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π (–¥–ª—è –∞–Ω–∞–ª–∏–∑–∞)
    probabilities = []
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–µ—Å—Ç–æ–≤–æ–º—É –æ–±—Ä–∞–∑—Ü—É
    for sample in X_test:
        # –î–æ–±–∞–≤–ª—è–µ–º 1 –≤ –Ω–∞—á–∞–ª–æ –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞ (intercept)
        sample_with_bias = np.concatenate([[1.0], sample])
        # –í—ã—á–∏—Å–ª—è–µ–º –ª–∏–Ω–µ–π–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
        z_value = np.dot(sample_with_bias, coefficients)
        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å –ø–æ–º–æ—â—å—é —Å–∏–≥–º–æ–∏–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
        probability = self.logistic_activation(z_value)
        # –ï—Å–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å >= –ø–æ—Ä–æ–≥–∞, –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º 1 (–±–æ–ª–µ–Ω), –∏–Ω–∞—á–µ 0 (–∑–¥–æ—Ä–æ–≤)
        predictions.append(1 if probability >= decision_threshold else 0)
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –∞–Ω–∞–ª–∏–∑–∞
        probabilities.append(probability)
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–Ø
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–ø–∏—Å–æ–∫ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ numpy –º–∞—Å—Å–∏–≤ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ —Ä–∞–±–æ—Ç—ã
    predictions_array = np.array(predictions)
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–ø–∏—Å–æ–∫ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –≤ numpy –º–∞—Å—Å–∏–≤
    probabilities_array = np.array(probabilities)
    
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± —É—Å–ø–µ—à–Ω–æ–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print("‚úì –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: {len(predictions_array)}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π –ø–æ—Ä–æ–≥ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print(f"‚úì –ü–æ—Ä–æ–≥ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏: {decision_threshold}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ä–∞–∑–¥–µ–ª–∞ —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π:")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∫–ª–∞—Å—Å–∞ 0 (–∑–¥–æ—Ä–æ–≤—ã–µ)
    print(f"  - –ö–ª–∞—Å—Å 0 (–∑–¥–æ—Ä–æ–≤—ã–µ): {(predictions_array == 0).sum()} ({np.mean(predictions_array == 0):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∫–ª–∞—Å—Å–∞ 1 (–±–æ–ª—å–Ω—ã–µ)
    print(f"  - –ö–ª–∞—Å—Å 1 (–±–æ–ª—å–Ω—ã–µ): {(predictions_array == 1).sum()} ({np.mean(predictions_array == 1):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –î–∏–∞–ø–∞–∑–æ–Ω –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π: [{probabilities_array.min():.3f}, {probabilities_array.max():.3f}]")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π
    print(f"‚úì –°—Ä–µ–¥–Ω—è—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å: {probabilities_array.mean():.3f}\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –º–∞—Å—Å–∏–≤ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏–∑ –º–µ—Ç–æ–¥–∞
    return predictions_array

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥ –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.predict_logistic = predict_logistic

# –í–´–ü–û–õ–ù–Ø–ï–ú –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ï –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–û–ô –†–ï–ì–†–ï–°–°–ò–ò
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 4: –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–ê–Ø –†–ï–ì–†–ï–°–°–ò–Ø - –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ï")
lr_predictions = classifier.predict_logistic(X_test, lr_coefficients)
print("‚úì –ó–ê–î–ê–ù–ò–ï 4 –í–´–ü–û–õ–ù–ï–ù–û!")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def identify_uncertain_predictions(self, X_test, coefficients, uncertainty_margin=0.25):
    """–ó–ê–î–ê–ù–ò–ï 5: –ü–æ–∏—Å–∫ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 5: –ê–ù–ê–õ–ò–ó –ù–ï–°–¢–ê–ë–ò–õ–¨–ù–´–• –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è –∏–Ω–¥–µ–∫—Å–æ–≤ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    uncertain_indices = []
    # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    uncertain_probabilities = []
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –≤—Å–µ–º —Ç–µ—Å—Ç–æ–≤—ã–º –æ–±—Ä–∞–∑—Ü–∞–º —Å –∏—Ö –∏–Ω–¥–µ–∫—Å–∞–º–∏ (enumerate –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏–Ω–¥–µ–∫—Å –∏ –∑–Ω–∞—á–µ–Ω–∏–µ)
    for idx, sample in enumerate(X_test):
        # –î–æ–±–∞–≤–ª—è–µ–º 1 –≤ –Ω–∞—á–∞–ª–æ –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞
        sample_with_bias = np.concatenate([[1.0], sample])
        # –í—ã—á–∏—Å–ª—è–µ–º –ª–∏–Ω–µ–π–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
        z_value = np.dot(sample_with_bias, coefficients)
        # –í—ã—á–∏—Å–ª—è–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å –ø–æ–º–æ—â—å—é —Å–∏–≥–º–æ–∏–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
        probability = self.logistic_activation(z_value)
        
        # –ï—Å–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –±–ª–∏–∑–∫–∞ –∫ 0.5 (–≤ –ø—Ä–µ–¥–µ–ª–∞—Ö uncertainty_margin)
        if abs(probability - 0.5) <= uncertainty_margin:
            # –î–æ–±–∞–≤–ª—è–µ–º –∏–Ω–¥–µ–∫—Å –≤ —Å–ø–∏—Å–æ–∫ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
            uncertain_indices.append(idx)
            # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
            uncertain_probabilities.append(probability)
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –ê–ù–ê–õ–ò–ó–ê –ù–ï–°–¢–ê–ë–ò–õ–¨–ù–û–°–¢–ò
    # –í—ã—á–∏—Å–ª—è–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–µ—Å—Ç–æ–≤—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤
    total_samples = len(X_test)
    # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    uncertain_count = len(uncertain_indices)
    # –í—ã—á–∏—Å–ª—è–µ–º –¥–æ–ª—é –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    uncertain_percentage = uncertain_count / total_samples
    
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –∞–Ω–∞–ª–∏–∑–∞
    print("‚úì –ê–Ω–∞–ª–∏–∑ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∑–∞–≤–µ—Ä—à–µ–Ω")
    # –ü–µ—á–∞—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–µ—Å—Ç–æ–≤—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤
    print(f"‚úì –í—Å–µ–≥–æ —Ç–µ—Å—Ç–æ–≤—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤: {total_samples}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –ù–∞–π–¥–µ–Ω–æ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: {uncertain_count}")
    # –ü–µ—á–∞—Ç–∞–µ–º –¥–æ–ª—é –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö
    print(f"‚úì –î–æ–ª—è –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö: {uncertain_percentage:.1%}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π –ø–æ—Ä–æ–≥ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏
    print(f"‚úì –ü–æ—Ä–æ–≥ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏: 0.5 ¬± {uncertainty_margin}")
    # –ü–µ—á–∞—Ç–∞–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å—Ä–µ–¥–∏ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –î–∏–∞–ø–∞–∑–æ–Ω –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π: [{min(uncertain_probabilities):.3f}, {max(uncertain_probabilities):.3f}]")
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —É—Å–ª–æ–≤–∏—è (–Ω–µ –º–µ–Ω–µ–µ 20% –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π)
    # –í—ã—á–∏—Å–ª—è–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ —Ç—Ä–µ–±—É–µ–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (20% –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞)
    min_required = int(total_samples * 0.2)
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —É—Å–ª–æ–≤–∏—è
    if uncertain_count >= min_required:
        # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ —á—Ç–æ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ
        print("‚úì ‚úÖ –£–°–õ–û–í–ò–ï –í–´–ü–û–õ–ù–ï–ù–û: –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π ‚â• 20%")
    else:
        # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ —á—Ç–æ —É—Å–ª–æ–≤–∏–µ –Ω–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ
        print("‚úì ‚ùå –£–°–õ–û–í–ò–ï –ù–ï –í–´–ü–û–õ–ù–ï–ù–û: –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π < 20%")
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –º–∞—Ä–∫–∏—Ä–æ–≤–∫–µ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤
    print("‚úì –ù–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã–µ –æ–±—Ä–∞–∑—Ü—ã –ø–æ–º–µ—á–µ–Ω—ã –º–µ—Ç–∫–æ–π -1\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ø–∏—Å–æ–∫ –∏–Ω–¥–µ–∫—Å–æ–≤ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏–∑ –º–µ—Ç–æ–¥–∞
    return uncertain_indices

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥ –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.identify_uncertain_predictions = identify_uncertain_predictions

# –í–´–ü–û–õ–ù–Ø–ï–ú –ê–ù–ê–õ–ò–ó –ù–ï–°–¢–ê–ë–ò–õ–¨–ù–´–• –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 5: –ê–ù–ê–õ–ò–ó –ù–ï–°–¢–ê–ë–ò–õ–¨–ù–´–• –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô")
uncertain_indices = classifier.identify_uncertain_predictions(X_test, lr_coefficients)
print("‚úì –ó–ê–î–ê–ù–ò–ï 5 –í–´–ü–û–õ–ù–ï–ù–û!")

# –°–æ–∑–¥–∞–µ–º —Ç–µ—Ä–Ω–∞—Ä–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
ternary_predictions = lr_predictions.copy()
for idx in uncertain_indices:
    ternary_predictions[idx] = -1
print(f"‚úì –°–æ–∑–¥–∞–Ω—ã —Ç–µ—Ä–Ω–∞—Ä–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è: {len(ternary_predictions)} –∑–∞–ø–∏—Å–µ–π")
print(f"‚úì –ù–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è: {(ternary_predictions == -1).sum()} –∑–∞–ø–∏—Å–µ–π")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def custom_distance(self, point1, point2):
    """–ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ"""
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞–∑–Ω–æ—Å—Ç–µ–π
    squared_distance = 0.0
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –≤—Å–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Ç–æ—á–µ–∫
    for i in range(len(point1)):
        # –°—É–º–º–∏—Ä—É–µ–º –∫–≤–∞–¥—Ä–∞—Ç—ã —Ä–∞–∑–Ω–æ—Å—Ç–µ–π –ø–æ –∫–∞–∂–¥–æ–π –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–µ
        squared_distance += (point1[i] - point2[i]) ** 2
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∫–æ—Ä–µ–Ω—å –∏–∑ —Å—É–º–º—ã –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ (–µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ)
    return math.sqrt(squared_distance)

def custom_sort(self, distances):
    """–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –º–µ—Ç–æ–¥–æ–º –≤—ã–±–æ—Ä–∞"""
    # –ü–æ–ª—É—á–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
    n = len(distances)
    # –í–Ω–µ—à–Ω–∏–π —Ü–∏–∫–ª - –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º
    for i in range(n):
        # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º —á—Ç–æ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π
        min_idx = i
        # –í–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ü–∏–∫–ª - –∏—â–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å—Ä–µ–¥–∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è
        for j in range(i + 1, n):
            # –ï—Å–ª–∏ –Ω–∞—à–ª–∏ —ç–ª–µ–º–µ–Ω—Ç –º–µ–Ω—å—à–µ —Ç–µ–∫—É—â–µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ (—Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è)
            if distances[j][0] < distances[min_idx][0]:
                # –û–±–Ω–æ–≤–ª—è–µ–º –∏–Ω–¥–µ–∫—Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
                min_idx = j
        # –ú–µ–Ω—è–µ–º –º–µ—Å—Ç–∞–º–∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –∏ –Ω–∞–π–¥–µ–Ω–Ω—ã–π –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π
        distances[i], distances[min_idx] = distances[min_idx], distances[i]
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫
    return distances

def knn_classify(self, X_train, y_train, X_test, neighbors_count=5):
    """–ó–ê–î–ê–ù–ò–ï 6: –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –º–µ—Ç–æ–¥–æ–º –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 6: –ú–ï–¢–û–î –ë–õ–ò–ñ–ê–ô–®–ò–• –°–û–°–ï–î–ï–ô")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    predictions = []
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –Ω–∞—á–∞–ª–µ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print(f"–ó–∞–ø—É—Å–∫ KNN –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏...")
    # –ü–µ—á–∞—Ç–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã KNN: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Å–µ–¥–µ–π, —Ä–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–æ–∫
    print(f"–ü–∞—Ä–∞–º–µ—Ç—Ä—ã: k={neighbors_count}, train_samples={len(X_train)}, test_samples={len(X_test)}")
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–µ—Å—Ç–æ–≤–æ–º—É –æ–±—Ä–∞–∑—Ü—É —Å –µ–≥–æ –∏–Ω–¥–µ–∫—Å–æ–º
    for i, test_sample in enumerate(X_test):
        # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π
        distances = []
        
        # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –¥–æ –≤—Å–µ—Ö –æ–±—É—á–∞—é—â–∏—Ö –æ–±—Ä–∞–∑—Ü–æ–≤
        # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É –æ–±—É—á–∞—é—â–µ–º—É –æ–±—Ä–∞–∑—Ü—É —Å –µ–≥–æ –∏–Ω–¥–µ–∫—Å–æ–º
        for j, train_sample in enumerate(X_train):
            # –í—ã—á–∏—Å–ª—è–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ç–µ—Å—Ç–æ–≤—ã–º –∏ —Ç–µ–∫—É—â–∏–º –æ–±—É—á–∞—é—â–∏–º –æ–±—Ä–∞–∑—Ü–æ–º
            dist = self.custom_distance(test_sample, train_sample)
            # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â—É—é –º–µ—Ç–∫—É –∫–ª–∞—Å—Å–∞ –≤ –≤–∏–¥–µ –∫–æ—Ä—Ç–µ–∂–∞
            distances.append((dist, y_train[j]))
        
        # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π
        # –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é —Å–ø–∏—Å–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –µ–≥–æ (—á—Ç–æ–±—ã –Ω–µ –∏–∑–º–µ–Ω—è—Ç—å –æ—Ä–∏–≥–∏–Ω–∞–ª)
        sorted_distances = self.custom_sort(distances.copy())
        
        # –í—ã–±–æ—Ä k –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π
        # –ë–µ—Ä–µ–º –ø–µ—Ä–≤—ã—Ö neighbors_count —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏–∑ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞
        nearest_neighbors = sorted_distances[:neighbors_count]
        
        # –ì–æ–ª–æ—Å–æ–≤–∞–Ω–∏–µ
        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ –≥–æ–ª–æ—Å–æ–≤ –¥–ª—è –∫–ª–∞—Å—Å–∞ 0
        class_0_votes = 0
        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ –≥–æ–ª–æ—Å–æ–≤ –¥–ª—è –∫–ª–∞—Å—Å–∞ 1
        class_1_votes = 0
        
        # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –≤—Å–µ–º –±–ª–∏–∂–∞–π—à–∏–º —Å–æ—Å–µ–¥—è–º
        for _, label in nearest_neighbors:
            # –ï—Å–ª–∏ –º–µ—Ç–∫–∞ –∫–ª–∞—Å—Å–∞ = 0, —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –∫–ª–∞—Å—Å–∞ 0
            if label == 0:
                class_0_votes += 1
            else:
                # –ò–Ω–∞—á–µ —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –∫–ª–∞—Å—Å–∞ 1
                class_1_votes += 1
        
        # –ü—Ä–∏–Ω—è—Ç–∏–µ —Ä–µ—à–µ–Ω–∏—è –ø–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤—É –≥–æ–ª–æ—Å–æ–≤
        # –ï—Å–ª–∏ –≥–æ–ª–æ—Å–æ–≤ –∑–∞ –∫–ª–∞—Å—Å 1 –±–æ–ª—å—à–µ, –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º 1, –∏–Ω–∞—á–µ 0
        prediction = 1 if class_1_votes > class_0_votes else 0
        # –î–æ–±–∞–≤–ª—è–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –≤ —Å–ø–∏—Å–æ–∫
        predictions.append(prediction)
        
        # –ü—Ä–æ–≥—Ä–µ—Å—Å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
        # –ö–∞–∂–¥—ã–µ 50 –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤ –≤—ã–≤–æ–¥–∏–º –ø—Ä–æ–≥—Ä–µ—Å—Å
        if (i + 1) % 50 == 0:
            # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤
            print(f"–û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i + 1}/{len(X_test)} —Ç–µ—Å—Ç–æ–≤—ã—Ö –æ–±—Ä–∞–∑—Ü–æ–≤")
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í KNN
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–ø–∏—Å–æ–∫ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ numpy –º–∞—Å—Å–∏–≤
    predictions_array = np.array(predictions)
    
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print("\n‚úì KNN –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –∑–∞–≤–µ—Ä—à–µ–Ω–∞!")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: {len(predictions_array)}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ä–∞–∑–¥–µ–ª–∞ —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"‚úì –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π KNN:")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∫–ª–∞—Å—Å–∞ 0
    print(f"  - –ö–ª–∞—Å—Å 0 (–∑–¥–æ—Ä–æ–≤—ã–µ): {(predictions_array == 0).sum()} ({np.mean(predictions_array == 0):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∫–ª–∞—Å—Å–∞ 1
    print(f"  - –ö–ª–∞—Å—Å 1 (–±–æ–ª—å–Ω—ã–µ): {(predictions_array == 1).sum()} ({np.mean(predictions_array == 1):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ k (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Å–µ–¥–µ–π)
    print(f"‚úì –ü–∞—Ä–∞–º–µ—Ç—Ä k: {neighbors_count}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–π –º–µ—Ç—Ä–∏–∫–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è
    print("‚úì –ú–µ—Ç—Ä–∏–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è: –ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ")
    # –ü–µ—á–∞—Ç–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–º –∞–ª–≥–æ—Ä–∏—Ç–º–µ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
    print("‚úì –ê–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏: –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –º–∞—Å—Å–∏–≤ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏–∑ –º–µ—Ç–æ–¥–∞
    return predictions_array

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥—ã –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.custom_distance = custom_distance
MedicalDataClassifier.custom_sort = custom_sort
MedicalDataClassifier.knn_classify = knn_classify

# –í–´–ü–û–õ–ù–Ø–ï–ú KNN –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–Æ
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 6: –ú–ï–¢–û–î –ë–õ–ò–ñ–ê–ô–®–ò–• –°–û–°–ï–î–ï–ô")
knn_predictions = classifier.knn_classify(X_train, y_train, X_test)
print("‚úì –ó–ê–î–ê–ù–ò–ï 6 –í–´–ü–û–õ–ù–ï–ù–û!")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def evaluate_performance(self, y_true, y_pred, method_name):
    """–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ –∫–∞—á–µ—Å—Ç–≤–∞ –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏"""
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫–∏ –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫
    # True Positive - –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –±–æ–ª—å–Ω—ã–µ
    tp = 0
    # False Positive - –∑–¥–æ—Ä–æ–≤—ã–µ, –æ—à–∏–±–æ—á–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–∞–∫ –±–æ–ª—å–Ω—ã–µ
    fp = 0
    # True Negative - –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–¥–æ—Ä–æ–≤—ã–µ
    tn = 0
    # False Negative - –±–æ–ª—å–Ω—ã–µ, –æ—à–∏–±–æ—á–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–∞–∫ –∑–¥–æ—Ä–æ–≤—ã–µ
    fn = 0
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –≤—Å–µ–º —Ä–µ–∞–ª—å–Ω—ã–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –º–µ—Ç–∫–∞–º –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ (zip –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç)
    for actual, predicted in zip(y_true, y_pred):
        # –°–ª—É—á–∞–π 1: —Ä–µ–∞–ª—å–Ω–æ –±–æ–ª—å–Ω–æ–π –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω –∫–∞–∫ –±–æ–ª—å–Ω–æ–π
        if actual == 1 and predicted == 1:
            # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ True Positive
            tp += 1
        # –°–ª—É—á–∞–π 2: —Ä–µ–∞–ª—å–Ω–æ –∑–¥–æ—Ä–æ–≤—ã–π –Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω –∫–∞–∫ –±–æ–ª—å–Ω–æ–π
        elif actual == 0 and predicted == 1:
            # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ False Positive
            fp += 1
        # –°–ª—É—á–∞–π 3: —Ä–µ–∞–ª—å–Ω–æ –∑–¥–æ—Ä–æ–≤—ã–π –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω –∫–∞–∫ –∑–¥–æ—Ä–æ–≤—ã–π
        elif actual == 0 and predicted == 0:
            # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ True Negative
            tn += 1
        # –°–ª—É—á–∞–π 4: —Ä–µ–∞–ª—å–Ω–æ –±–æ–ª—å–Ω–æ–π –Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω –∫–∞–∫ –∑–¥–æ—Ä–æ–≤—ã–π
        elif actual == 1 and predicted == 0:
            # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ False Negative
            fn += 1
    
    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫ –∫–∞—á–µ—Å—Ç–≤–∞
    # Precision = TP / (TP + FP) - —Ç–æ—á–Ω–æ—Å—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–≥–Ω–æ–∑–æ–≤
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    # Recall = TP / (TP + FN) - –ø–æ–ª–Ω–æ—Ç–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –±–æ–ª—å–Ω—ã—Ö
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    # F1-score = 2 * (precision * recall) / (precision + recall) - –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ
    f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    # Accuracy = (TP + TN) / total - –æ–±—â–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    accuracy = (tp + tn) / len(y_true)
    
    # –í–´–í–û–î –ú–ï–¢–†–ò–ö –ö–ê–ß–ï–°–¢–í–ê
    # –ù–∞–∑–≤–∞–Ω–∏–µ –º–µ—Ç–æ–¥–∞
    print(f"–ú–ï–¢–†–ò–ö–ò –ö–ê–ß–ï–°–¢–í–ê ({method_name}):")
    # –û–±—â–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å —Å —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ–º –¥–æ 3 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
    print(f"  ‚Ä¢ Accuracy:  {accuracy:.3f}")
    # –¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    print(f"  ‚Ä¢ Precision: {precision:.3f}")
    # –ü–æ–ª–Ω–æ—Ç–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è
    print(f"  ‚Ä¢ Recall:    {recall:.3f}")
    # F1-score (–°–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞)
    print(f"  ‚Ä¢ F1-score:  {f1_score:.3f}")
    # –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
    print(f"  ‚Ä¢ –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫: TP={tp}, FP={fp}, TN={tn}, FN={fn}")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –≤—Å–µ –≤—ã—á–∏—Å–ª–µ–Ω–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫
    return precision, recall, f1_score, accuracy, tp, fp, tn, fn

def evaluate_ternary_classification(self, y_true, ternary_predictions):
    """–û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏ –¥–ª—è —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏"""
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    correct_predictions = 0
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π (–Ω–µ -1)
    total_stable = 0
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ –≤—Å–µ–º —Ä–µ–∞–ª—å–Ω—ã–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –º–µ—Ç–∫–∞–º
    for actual, predicted in zip(y_true, ternary_predictions):
        # –£—á–∏—Ç—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ —Å—Ç–∞–±–∏–ª—å–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è (–Ω–µ -1)
        if predicted != -1:
            # –ï—Å–ª–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ
            if actual == predicted:
                # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
                correct_predictions += 1
            # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
            total_stable += 1
    
    # –í—ã—á–∏—Å–ª—è–µ–º accuracy —Ç–æ–ª—å–∫–æ –¥–ª—è —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    accuracy = correct_predictions / total_stable if total_stable > 0 else 0
    
    # –ü–µ—á–∞—Ç–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print(f"–ú–ï–¢–†–ò–ö–ò –ö–ê–ß–ï–°–¢–í–ê (–¢–µ—Ä–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è):")
    # –ü–µ—á–∞—Ç–∞–µ–º accuracy –¥–ª—è —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"  ‚Ä¢ Accuracy (—Ç–æ–ª—å–∫–æ —Å—Ç–∞–±–∏–ª—å–Ω—ã–µ): {accuracy:.3f}")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –∏—Ö –¥–æ–ª—é
    print(f"  ‚Ä¢ –°—Ç–∞–±–∏–ª—å–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è: {total_stable}/{len(y_true)} ({total_stable/len(y_true):.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –∏—Ö –¥–æ–ª—é
    print(f"  ‚Ä¢ –ù–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è: {len(y_true) - total_stable}/{len(y_true)} ({(len(y_true) - total_stable)/len(y_true):.1%})")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º accuracy
    return accuracy

def run_quality_evaluation(self, y_test, lr_predictions, knn_predictions, ternary_predictions):
    """–ó–ê–î–ê–ù–ò–ï 7: –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 7: –û–¶–ï–ù–ö–ê –ö–ê–ß–ï–°–¢–í–ê –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–ò")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –Ω–∞—á–∞–ª–µ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞
    print("–ü—Ä–æ–≤–æ–¥–∏–º –æ—Ü–µ–Ω–∫—É –∫–∞—á–µ—Å—Ç–≤–∞ –≤—Å–µ—Ö –º–µ—Ç–æ–¥–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏...\n")
    
    # –û—Ü–µ–Ω–∫–∞ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    print("1. –õ–û–ì–ò–°–¢–ò–ß–ï–°–ö–ê–Ø –†–ï–ì–†–ï–°–°–ò–Ø (–±–∏–Ω–∞—Ä–Ω–∞—è):")
    # –í—ã—á–∏—Å–ª—è–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    lr_precision, lr_recall, lr_f1, lr_accuracy, tp_lr, fp_lr, tn_lr, fn_lr = self.evaluate_performance(
        y_test, lr_predictions, "–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è")
    # –ü–µ—á–∞—Ç–∞–µ–º –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –≤–∏–∑—É–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è
    print()
    
    # –û—Ü–µ–Ω–∫–∞ KNN
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è KNN
    print("2. –ú–ï–¢–û–î –ë–õ–ò–ñ–ê–ô–®–ò–• –°–û–°–ï–î–ï–ô (–±–∏–Ω–∞—Ä–Ω–∞—è):")
    # –í—ã—á–∏—Å–ª—è–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è KNN
    knn_precision, knn_recall, knn_f1, knn_accuracy, tp_knn, fp_knn, tn_knn, fn_knn = self.evaluate_performance(
        y_test, knn_predictions, "KNN")
    # –ü–µ—á–∞—Ç–∞–µ–º –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –≤–∏–∑—É–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è
    print()
    
    # –û—Ü–µ–Ω–∫–∞ —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print("3. –¢–ï–†–ù–ê–†–ù–ê–Ø –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–Ø:")
    # –í—ã—á–∏—Å–ª—è–µ–º accuracy –¥–ª—è —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    ternary_accuracy = self.evaluate_ternary_classification(y_test, ternary_predictions)
    # –ü–µ—á–∞—Ç–∞–µ–º –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –≤–∏–∑—É–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è
    print()
    
    # –°–†–ê–í–ù–ò–¢–ï–õ–¨–ù–ê–Ø –¢–ê–ë–õ–ò–¶–ê –ú–ï–¢–†–ò–ö
    # –ü–µ—á–∞—Ç–∞–µ–º –≤–µ—Ä—Ö–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É —Ç–∞–±–ª–∏—Ü—ã –∏–∑ 70 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 70)
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Ç–∞–±–ª–∏—Ü—ã
    print("–°–†–ê–í–ù–ò–¢–ï–õ–¨–ù–ê–Ø –¢–ê–ë–õ–ò–¶–ê –ú–ï–¢–†–ò–ö")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 70 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 70)
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–∫–∏ —Å—Ç–æ–ª–±—Ü–æ–≤ —Ç–∞–±–ª–∏—Ü—ã —Å –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏–µ–º –ø–æ –ª–µ–≤–æ–º—É –∫—Ä–∞—é
    print(f"{'–ú–ï–¢–†–ò–ö–ê':<12} {'–õ–û–ì–ò–°–¢–ò–ö–ê':<12} {'KNN':<12} {'–¢–ï–†–ù–ê–†–ù–ê–Ø':<12}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 70 –∑–Ω–∞–∫–æ–≤ "-"
    print("-" * 70)
    # –ü–µ—á–∞—Ç–∞–µ–º —Å—Ç—Ä–æ–∫—É —Å precision –¥–ª—è –≤—Å–µ—Ö –º–µ—Ç–æ–¥–æ–≤
    print(f"{'Precision':<12} {lr_precision:<12.3f} {knn_precision:<12.3f} {'-':<12}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å—Ç—Ä–æ–∫—É —Å recall –¥–ª—è –≤—Å–µ—Ö –º–µ—Ç–æ–¥–æ–≤
    print(f"{'Recall':<12} {lr_recall:<12.3f} {knn_recall:<12.3f} {'-':<12}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å—Ç—Ä–æ–∫—É —Å F1-score –¥–ª—è –≤—Å–µ—Ö –º–µ—Ç–æ–¥–æ–≤
    print(f"{'F1-score':<12} {lr_f1:<12.3f} {knn_f1:<12.3f} {'-':<12}")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å—Ç—Ä–æ–∫—É —Å accuracy –¥–ª—è –≤—Å–µ—Ö –º–µ—Ç–æ–¥–æ–≤
    print(f"{'Accuracy':<12} {lr_accuracy:<12.3f} {knn_accuracy:<12.3f} {ternary_accuracy:<12.3f}")
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∏–∂–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É —Ç–∞–±–ª–∏—Ü—ã –∏–∑ 70 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 70)
    
    # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –û–¶–ï–ù–ö–ò –ö–ê–ß–ï–°–¢–í–ê
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞
    print("\n‚úì –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∑–∞–≤–µ—Ä—à–µ–Ω–∞!")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± —É—Å–ø–µ—à–Ω–æ–º –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫
    print("‚úì –í—Å–µ –º–µ—Ç—Ä–∏–∫–∏ —É—Å–ø–µ—à–Ω–æ –≤—ã—á–∏—Å–ª–µ–Ω—ã")
    # –ü–µ—á–∞—Ç–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –ø—Ä–æ–≤–µ–¥–µ–Ω–Ω–æ–º —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω–æ–º –∞–Ω–∞–ª–∏–∑–µ
    print("‚úì –°—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –ø—Ä–æ–≤–µ–¥–µ–Ω\n")
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏ –≤—Å–µ—Ö –º–µ—Ç–æ–¥–æ–≤ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
    return {
        'logistic': (lr_precision, lr_recall, lr_f1, lr_accuracy),
        'knn': (knn_precision, knn_recall, knn_f1, knn_accuracy),
        'ternary': ternary_accuracy
    }

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥—ã –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.evaluate_performance = evaluate_performance
MedicalDataClassifier.evaluate_ternary_classification = evaluate_ternary_classification
MedicalDataClassifier.run_quality_evaluation = run_quality_evaluation

# –í–´–ü–û–õ–ù–Ø–ï–ú –û–¶–ï–ù–ö–£ –ö–ê–ß–ï–°–¢–í–ê
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 7: –û–¶–ï–ù–ö–ê –ö–ê–ß–ï–°–¢–í–ê –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–ò")
metrics = classifier.run_quality_evaluation(y_test, lr_predictions, knn_predictions, ternary_predictions)
print("‚úì –ó–ê–î–ê–ù–ò–ï 7 –í–´–ü–û–õ–ù–ï–ù–û!")
```

```python
# –ü—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ MedicalDataClassifier
def interpret_results(self, metrics, uncertain_count, total_samples):
    """–ó–ê–î–ê–ù–ò–ï 8: –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤"""
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª–æ–∫–∞
    print("–ë–õ–û–ö 8: –ò–ù–¢–ï–†–ü–†–ï–¢–ê–¢–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)
    
    # –ò–∑–≤–ª–µ–∫–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏–∑ —Å–ª–æ–≤–∞—Ä—è
    lr_precision, lr_recall, lr_f1, lr_accuracy = metrics['logistic']
    # –ò–∑–≤–ª–µ–∫–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ KNN –∏–∑ —Å–ª–æ–≤–∞—Ä—è
    knn_precision, knn_recall, knn_f1, knn_accuracy = metrics['knn']
    # –ò–∑–≤–ª–µ–∫–∞–µ–º accuracy —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏–∑ —Å–ª–æ–≤–∞—Ä—è
    ternary_accuracy = metrics['ternary']
    
    # –ü–µ—á–∞—Ç–∞–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    print("–ê–ù–ê–õ–ò–ó –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –ò –í–´–í–û–î–´:\n")
    
    # –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print("1. –°–†–ê–í–ù–ï–ù–ò–ï –ú–ï–¢–û–î–û–í –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–ò:")
    # –ê–Ω–∞–ª–∏–∑ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    print("   ‚Ä¢ –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –ø–æ–∫–∞–∑–∞–ª–∞:")
    # –ü–µ—á–∞—Ç–∞–µ–º accuracy —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - Accuracy: {lr_accuracy:.3f} - —Ö–æ—Ä–æ—à–∞—è –æ–±—â–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å")
    # –ü–µ—á–∞—Ç–∞–µ–º precision —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - Precision: {lr_precision:.3f} - –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–≥–Ω–æ–∑–æ–≤")
    # –ü–µ—á–∞—Ç–∞–µ–º recall —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - Recall: {lr_recall:.3f} - —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –Ω–∞—Ö–æ–¥–∏—Ç—å –±–æ–ª—å–Ω—ã—Ö –ø–∞—Ü–∏–µ–Ω—Ç–æ–≤")
    # –ü–µ—á–∞—Ç–∞–µ–º F1-score —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - F1-score: {lr_f1:.3f} - —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç—å —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ –ø–æ–ª–Ω–æ—Ç—ã")
    
    # –ê–Ω–∞–ª–∏–∑ KNN
    print("   ‚Ä¢ –ú–µ—Ç–æ–¥ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π (KNN):")
    # –ü–µ—á–∞—Ç–∞–µ–º accuracy KNN —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - Accuracy: {knn_accuracy:.3f} - —Å–æ–ø–æ—Å—Ç–∞–≤–∏–º–∞—è —Å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π")
    # –ü–µ—á–∞—Ç–∞–µ–º precision KNN —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - Precision: {knn_precision:.3f} - –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–≥–Ω–æ–∑–æ–≤")
    # –ü–µ—á–∞—Ç–∞–µ–º recall KNN —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–º
    print(f"     - Recall: {knn_recall:.3f} - –ø–æ–ª–Ω–æ—Ç–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏–π")
    # –ü–µ—á–∞—Ç–∞–µ–º –∞–±—Å–æ–ª—é—Ç–Ω—É—é —Ä–∞–∑–Ω–∏—Ü—É –≤ accuracy –º–µ–∂–¥—É –º–µ—Ç–æ–¥–∞–º–∏
    print(f"     - –†–∞–∑–Ω–∏—Ü–∞ –≤ accuracy: {abs(lr_accuracy - knn_accuracy):.3f}")
    
    # –ê–Ω–∞–ª–∏–∑ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print("\n2. –ê–ù–ê–õ–ò–ó –ù–ï–°–¢–ê–ë–ò–õ–¨–ù–´–• –ü–†–ï–î–°–ö–ê–ó–ê–ù–ò–ô:")
    # –ü–µ—á–∞—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    print(f"   ‚Ä¢ –û–±–Ω–∞—Ä—É–∂–µ–Ω–æ {uncertain_count} –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π ({uncertain_count/total_samples:.1%})")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—é –¥–ª—è –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤
    print("   ‚Ä¢ –≠—Ç–∏ —Å–ª—É—á–∞–∏ —Ç—Ä–µ–±—É—é—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–≥–æ –º–µ–¥–∏—Ü–∏–Ω—Å–∫–æ–≥–æ –æ–±—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è")
    # –ü–µ—á–∞—Ç–∞–µ–º –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print("   ‚Ä¢ –¢–µ—Ä–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã–¥–µ–ª–∏—Ç—å '–≥—Ä—É–ø–ø—É —Ä–∏—Å–∫–∞'")
    # –ü–µ—á–∞—Ç–∞–µ–º accuracy –Ω–∞ —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è—Ö
    print(f"   ‚Ä¢ Accuracy –Ω–∞ —Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è—Ö: {ternary_accuracy:.3f}")
    
    # –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏
    print("\n3. –ü–†–ê–ö–¢–ò–ß–ï–°–ö–ò–ï –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò:")
    # –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è –ø–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—é –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    print("   ‚Ä¢ –î–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–π –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏ —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è")
    # –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è –ø–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—é KNN
    print("   ‚Ä¢ KNN –º–æ–∂–µ—Ç –±—ã—Ç—å –ø–æ–ª–µ–∑–µ–Ω –∫–∞–∫ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –º–µ—Ç–æ–¥ –ø—Ä–æ–≤–µ—Ä–∫–∏")
    # –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è –¥–ª—è –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤
    print("   ‚Ä¢ –ù–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã–µ —Å–ª—É—á–∞–∏ –¥–æ–ª–∂–Ω—ã –Ω–∞–ø—Ä–∞–≤–ª—è—Ç—å—Å—è –∫ –∫–∞—Ä–¥–∏–æ–ª–æ–≥—É")
    # –û–±—â–∞—è –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–µ–π
    print("   ‚Ä¢ –ú–æ–¥–µ–ª–∏ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—Ç –∫–ª–∏–Ω–∏—á–µ—Å–∫–∏ –ø—Ä–∏–º–µ–Ω–∏–º–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ")
    
    # –ò—Ç–æ–≥–æ–≤—ã–µ –≤—ã–≤–æ–¥—ã
    print("\n4. –í–´–í–û–î–´:")
    # –í—ã–≤–æ–¥ 1: –∫–∞—á–µ—Å—Ç–≤–æ –º–µ—Ç–æ–¥–æ–≤
    print("   ‚úì –û–±–∞ –º–µ—Ç–æ–¥–∞ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç —Ö–æ—Ä–æ—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏")
    # –í—ã–≤–æ–¥ 2: —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤
    print("   ‚úì –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –Ω–µ–º–Ω–æ–≥–æ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç KNN –ø–æ —Ç–æ—á–Ω–æ—Å—Ç–∏")
    # –í—ã–≤–æ–¥ 3: –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    print("   ‚úì –í—ã–¥–µ–ª–µ–Ω–∏–µ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π —É–ª—É—á—à–∞–µ—Ç –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å —Å–∏—Å—Ç–µ–º—ã")
    # –í—ã–≤–æ–¥ 4: –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—å —Å–∏—Å—Ç–µ–º—ã
    print("   ‚úì –†–∞–∑—Ä–∞–±–æ—Ç–∞–Ω–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ –≥–æ—Ç–æ–≤–∞ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≤ –∫–ª–∏–Ω–∏—á–µ—Å–∫–∏—Ö —É—Å–ª–æ–≤–∏—è—Ö")
    
    # –ó–∞–≤–µ—Ä—à–∞—é—â–∏–µ —Å–æ–æ–±—â–µ–Ω–∏—è
    # –ü–µ—á–∞—Ç–∞–µ–º –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è
    print("\n" + "=" * 60)
    # –ü–µ—á–∞—Ç–∞–µ–º —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± —É—Å–ø–µ—à–Ω–æ–º –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –∞–Ω–∞–ª–∏–∑–∞
    print("–ê–ù–ê–õ–ò–ó –ó–ê–í–ï–†–®–ï–ù –£–°–ü–ï–®–ù–û!")
    # –ü–µ—á–∞—Ç–∞–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å–Ω—É—é –ª–∏–Ω–∏—é –∏–∑ 60 –∑–Ω–∞–∫–æ–≤ "="
    print("=" * 60)

# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥ –≤ –∫–ª–∞—Å—Å
MedicalDataClassifier.interpret_results = interpret_results

# –í–´–ü–û–õ–ù–Ø–ï–ú –ò–ù–¢–ï–†–ü–†–ï–¢–ê–¶–ò–Æ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í
print("üöÄ –í–´–ü–û–õ–ù–Ø–ï–ú –ó–ê–î–ê–ù–ò–ï 8: –ò–ù–¢–ï–†–ü–†–ï–¢–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
classifier.interpret_results(metrics, len(uncertain_indices), len(y_test))
print("‚úì –ó–ê–î–ê–ù–ò–ï 8 –í–´–ü–û–õ–ù–ï–ù–û!")
```
